````python
import pandas as pd
import numpy as np

# Generar dataset con 10 columnas numéricas y 100 filas
np.random.seed(123)  # para reproducibilidad
data = np.random.rand(100, 10)
column_names = [f'col_{i+1}' for i in range(10)]
df = pd.DataFrame(data, columns=column_names)

# Guardar el dataset en un archivo Excel
excel_file = '/content/dataset.xlsx'
df.to_excel(excel_file, index=False)

# Cargar el archivo Excel
df_loaded = pd.read_excel(excel_file)

# Calcular 15 medidas estadísticas para cada columna
stats = df_loaded.describe().T
stats['var'] = df_loaded.var()  # varianza
stats['skew'] = df_loaded.skew()  # skewness (asimetría)
stats['kurt'] = df_loaded.kurt()  # kurtosis (curtosis)
stats['range'] = df_loaded.max() - df_loaded.min()  # rango
stats['q1'] = df_loaded.quantile(0.25)  # primer cuartil
stats['q3'] = df_loaded.quantile(0.75)  # tercer cuartil
stats['iqr'] = stats['q3'] - stats['q1']  # rango intercuartil

# Selección de medidas estadísticas
selected_stats = stats[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'var', 'skew', 'kurt', 'range', 'q1', 'q3', 'iqr']]

# Guardar las estadísticas en un nuevo archivo Excel
stats_excel_file = '/content/stats.xlsx'
selected_stats.to_excel(stats_excel_file)

# Mostrar las estadísticas calculadas
print(selected_stats)
````

# Estadística descriptiva básica

1. **Media (Promedio)**
$$
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$
   - **Interpretación:** La media representa el valor promedio en el conjunto de datos, es decir, la suma de todos los valores dividida por el número de datos.
   - **Donde:**
     - $\bar{x}$: Media.
     - $n$: Número de datos en el conjunto.
     - $x_i$: Valor individual en el conjunto.

2. **Mediana (Valor central)**
   - Si $n$ es impar:
   $$
   \text{Mediana} = x_{\frac{n+1}{2}}
   $$
   - Si $n$ es par:
   $$
   \text{Mediana} = \frac{x_{\frac{n}{2}} + x_{\frac{n}{2}+1}}{2}
   $$
   - **Interpretación:** La mediana es el valor que separa los datos en dos partes iguales; el 50% de los datos están por encima y el 50% están por debajo.
   - **Donde:**
     - $\text{Mediana}$: Mediana.
     - $n$: Número de datos en el conjunto.
     - $x_{i}$: Valor individual en el conjunto.

3. **Varianza (Dispersión de datos)**
$$
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
$$
   - **Interpretación:** La varianza mide la dispersión de los datos en relación con la media; cuanto mayor es la varianza, mayor es la dispersión de los datos.
   - **Donde:**
     - $\sigma^2$: Varianza.
     - $n$: Número de datos en el conjunto.
     - $x_i$: Valor individual en el conjunto.
     - $\bar{x}$: Media.

4. **Desviación Estándar (Raíz cuadrada de la varianza):**
$$
\sigma = \sqrt{\sigma^2}
$$
   - **Interpretación:** La desviación estándar es una medida de la dispersión que tiene la misma unidad que los datos originales. Es una versión más interpretable de la varianza.
   - **Donde:**
     - $\sigma$: Desviación Estándar.
     - $\sigma^2$: Varianza.

5. **Mínimo (Valor más pequeño en el dataset)**
$$
\text{Mínimo} = \min(x_i)
$$
   - **Interpretación:** El mínimo es el valor más pequeño en el conjunto de datos.
   - **Donde:**
     - $\text{Mínimo}$: Valor mínimo.
     - $x_i$: Valor individual en el conjunto.

6. **Máximo (Valor más grande en el dataset)**
$$
\text{Máximo} = \max(x_i)
$$
   - **Interpretación:** El máximo es el valor más grande en el conjunto de datos.
   - **Donde:**
     - $\text{Máximo}$: Valor máximo.
     - $x_i$: Valor individual en el conjunto.

7. **Rango (Diferencia entre el máximo y el mínimo)**
$$
\text{Rango} = \text{Máximo} - \text{Mínimo}
$$
   - **Interpretación:** El rango es la diferencia entre el valor más grande y el más pequeño en el conjunto de datos, lo que indica la amplitud total de los datos.
   - **Donde:**
     - $\text{Rango}$: Rango.
     - $\text{Máximo}$: Valor máximo.
     - $\text{Mínimo}$: Valor mínimo.

8. **Percentil 25 (Primer cuartil)**
   - El valor por debajo del cual se encuentra el 25% de los datos ordenados.

9. **Percentil 75 (Tercer cuartil)**
   - El valor por debajo del cual se encuentra el 75% de los datos ordenados.

10. **Rango Intercuartílico (Diferencia entre el tercer y el primer cuartil):**
$$
\text{Rango Intercuartílico} = \text{Percentil 75} - \text{Percentil 25}
$$
   - **Interpretación:** El rango intercuartílico mide la dispersión entre el primer cuartil (25%) y el tercer cuartil (75%) de los datos.

11. **Coeficiente de Variación (Relación entre la desviación estándar y la media):**
$$
\text{Coeficiente de Variación} = \left(\frac{\sigma}{\bar{x}}\right) \times 100\%
$$
   - **Interpretación:** El coeficiente de variación mide la variabilidad relativa en términos porcentuales. Cuanto mayor sea, mayor será la variabilidad en relación con la media.
   - **Donde:**
     - $\text{Coeficiente de Variación}$: Coeficiente de Variación.
     - $\sigma$: Desviación Estándar.
     - $\bar{x}$: Media.

12. **Skewness (Asimetría)**
$$
\text{Skewness} = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{x_i - \bar{x}}{\sigma}\right)^3
$$
   - **Interpretación:** La skewness mide la asimetría de la distribución de datos. Un valor positivo indica una cola larga a la derecha, mientras que un valor negativo indica una cola larga a la izquierda.
   - **Donde:**
     - $\text{Skewness}$: Skewness.
     - $n$: Número de datos en el conjunto.
     - $x_i$: Valor individual en el conjunto.
     - $\bar{x}$: Media.
     - $\sigma$: Desviación Estándar.

13. **Kurtosis (Curtosis)**
$$
\text{Kurtosis} = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{x_i - \bar{x}}{\sigma}\right)^4 - 3
$$
   - **Interpretación:** La kurtosis mide la forma de la distribución en relación con una distribución normal. Un valor positivo indica una distribución más puntiaguda que una normal, mientras que un valor negativo indica una distribución más achatada.
   - **Donde:**
     - $\text{Kurtosis}$: Kurtosis.
     - $n$: Número de datos en el conjunto.
     - $x_i$: Valor individual en el conjunto.
     - $\bar{x}$: Media.
     - $\sigma$: Desviación Estándar.

14. **Entropía (Medida de incertidumbre)**
$$
\text{Entropía} = -\sum_{i=1}^{n} P(x_i) \log_2(P(x_i))
$$
   - **Interpretación:** La entropía mide la incertidumbre en la distribución de datos. Cuanto mayor es la entropía, mayor es la incertidumbre.
   - **Donde:**
     - $\text{Entropía}$: Entropía.
     - $n$: Número de datos en el conjunto.
     - $P(x_i)$: Probabilidad de ocurrencia de $x_i$ en el conjunto.

15. **Coeficiente de Asimetría (Medida de asimetría en la distribución)**
$$
\text{Coeficiente de Asimetría} = \frac{3(\bar{x} - \text{Mediana})}{\sigma}
$$
   - **Interpretación:** El coeficiente de asimetría mide la asimetría de la distribución en relación con la mediana. Un valor positivo indica una asimetría positiva (cola larga a la derecha), mientras que un valor negativo indica una asimetría negativa (cola larga a la izquierda).
   - **Donde:**
     - $\text{Coeficiente de Asimetría}$: Coeficiente de Asimetría.
     - $\bar{x}$: Media.
     - $\text{Mediana}$: Mediana.
     - $\sigma$: Desviación Estándar.

# Estadística descriptiva básica en Python

Este código utiliza la biblioteca NumPy en Python para calcular una serie de estadísticas descriptivas a partir de un conjunto de datos aleatorios.

### Importar NumPy

El código comienza importando la biblioteca NumPy bajo el alias "np". NumPy es una biblioteca en Python que proporciona soporte para trabajar con matrices y funciones matemáticas.

```python
import numpy as np
```

### Generación de un conjunto de datos aleatorios

A continuación, el código genera un conjunto de datos de 1000 números aleatorios entre 0 y 100. Para que los números aleatorios sean reproducibles (es decir, para que obtengas los mismos números cada vez que ejecutes el código), se establece una semilla aleatoria con `np.random.seed(42)`.

```python
np.random.seed(42)
dataset = np.random.randint(0, 100, 1000)
```

- `np.random.seed(42)`: Esto fija la semilla del generador de números aleatorios de NumPy en 42.
- `np.random.randint(0, 100, 1000)`: Esto genera una matriz de 1000 números enteros aleatorios en el rango de 0 a 99 y la almacena en la variable `dataset`.

### Cálculo de estadísticas descriptivas

A continuación, el código calcula diversas estadísticas descriptivas basadas en el conjunto de datos generado. Estas estadísticas incluyen la media, la mediana, la varianza, la desviación estándar, el valor mínimo, el valor máximo, el rango, percentiles, coeficiente de variación, skewness, kurtosis, entropía, coeficiente de asimetría y coeficiente de curtosis.

Vamos a explicar cada uno de estos cálculos:

#### Media

```python
media = np.mean(dataset)
```

La línea anterior calcula la media (o promedio) de los datos en `dataset` utilizando la función `np.mean()` y la almacena en la variable `media`.

#### Mediana

```python
mediana = np.median(dataset)
```

Esta línea calcula la mediana de los datos en `dataset` utilizando la función `np.median()` y la almacena en la variable `mediana`. La mediana es el valor que se encuentra en el centro de un conjunto de datos ordenados.

#### Varianza y Desviación Estándar

```python
varianza = np.var(dataset)
desviacion_estandar = np.std(dataset)
```

Estas líneas calculan la varianza y la desviación estándar de los datos en `dataset` utilizando las funciones `np.var()` y `np.std()` respectivamente. La varianza mide cuánto se dispersan los datos y la desviación estándar es la raíz cuadrada de la varianza, que proporciona una medida de la dispersión en la misma escala que los datos originales.

#### Valor Mínimo y Valor Máximo

```python
minimo = np.min(dataset)
maximo = np.max(dataset)
```

Estas líneas calculan el valor mínimo y el valor máximo en `dataset` utilizando las funciones `np.min()` y `np.max()` respectivamente.

#### Rango

```python
rango = np.ptp(dataset)
```

Esta línea calcula el rango de los datos en `dataset` utilizando la función `np.ptp()`, que significa "punto a punto" y devuelve la diferencia entre el máximo y el mínimo.

#### Percentiles

```python
percentil_25 = np.percentile(dataset, 25)
percentil_75 = np.percentile(dataset, 75)
```

Estas líneas calculan los percentiles 25 y 75 de los datos en `dataset` utilizando la función `np.percentile()`. Los percentiles son valores que dividen los datos en porcentajes específicos, en este caso, el percentil 25 divide los datos en el 25% inferior y el percentil 75 divide los datos en el 75% inferior.

#### Rango Intercuartílico

```python
rango_intercuartilico = percentil_75 - percentil_25
```

Esta línea calcula el rango intercuartílico, que es la diferencia entre el percentil 75 y el percentil 25. Es una medida de la dispersión en la mitad central de los datos.

#### Coeficiente de Variación

```python
coeficiente_de_variacion = (desviacion_estandar / media) * 100
```

Esta línea calcula el coeficiente de variación, que es una medida de la variabilidad relativa de los datos. Se calcula como el cociente de la desviación estándar y la media, multiplicado por 100 para expresarlo en términos de porcentaje.

#### Skewness (Asimetría)

```python
skewness = np.mean(((dataset - media) / desviacion_estandar) ** 3)
```

Esta línea calcula el skewness, que mide la asimetría de la distribución de los datos. Un valor positivo indica una cola derecha (sesgo positivo), mientras que un valor negativo indica una cola izquierda (sesgo negativo).

#### Kurtosis

```python
kurtosis = np.mean(((dataset - media) / desviacion_estandar) ** 4) - 3
```

La línea anterior calcula la kurtosis, que mide la concentración de los datos alrededor de la media. Restando 3 a este valor, se obtiene la kurtosis de exceso, que se compara con 0 para determinar si la distribución es más puntiaguda (kurtosis positiva) o más aplanada (kurtosis negativa) que la distribución normal.

#### Entropía

```python
entropia = -np.sum(np.nan_to_num((np.bincount(dataset) / len(dataset)) * np.log2(np.bincount(dataset) / len(dataset)), nan=0))
```

Esta línea calcula la entropía de los datos. La entropía mide la incertidumbre o la información contenida en una distribución de probabilidad. En este caso, se calcula utilizando la fórmula de la entropía de Shannon.

- `np.bincount(dataset)`: Cuenta la frecuencia de cada valor único en `dataset`.
- `np.bincount(dataset) / len(dataset)`: Calcula la probabilidad de cada valor único en `dataset`.
- `np.log2()`: Calcula el logaritmo base 2 de cada probabilidad.
- `np.nan_to_num()`: Reemplaza los valores NaN (que podrían ocurrir si un valor único no está presente en `dataset`) con 0.
- `-np.sum()`: Suma los valores resultantes y cambia el signo

 para obtener la entropía.

#### Coeficiente de Asimetría

```python
coeficiente_de_asimetria = (3 * (media - mediana)) / desviacion_estandar
```

Esta línea calcula el coeficiente de asimetría, que mide la asimetría de la distribución. Se basa en la diferencia entre la media y la mediana, multiplicada por 3 y dividida por la desviación estándar.

#### Coeficiente de Curtosis

```python
coeficiente_de_curtosis = np.mean(((dataset - media) / desviacion_estandar) ** 4)
```

Esta línea calcula el coeficiente de curtosis, que mide la forma de la distribución de los datos. Cuanto mayor sea el valor, más puntiaguda será la distribución (picos altos) en comparación con una distribución normal.

Estas estadísticas proporcionan información importante sobre la distribución y las propiedades de los datos.

````python
import numpy as np

# Generamos un dataset de 1000 números aleatorios entre 0 y 100
np.random.seed(42)
dataset = np.random.randint(0, 100, 1000)

# Calculamos las estadísticas descriptivas
media = np.mean(dataset)
mediana = np.median(dataset)
varianza = np.var(dataset)
desviacion_estandar = np.std(dataset)
minimo = np.min(dataset)
maximo = np.max(dataset)
rango = np.ptp(dataset)
percentil_25 = np.percentile(dataset, 25)
percentil_75 = np.percentile(dataset, 75)
rango_intercuartilico = percentil_75 - percentil_25
coeficiente_de_variacion = (desviacion_estandar / media) * 100
skewness = np.mean(((dataset - media) / desviacion_estandar) ** 3)
kurtosis = np.mean(((dataset - media) / desviacion_estandar) ** 4) - 3
entropia = -np.sum(np.nan_to_num((np.bincount(dataset) / len(dataset)) * np.log2(np.bincount(dataset) / len(dataset)), nan=0))
coeficiente_de_asimetria = (3 * (media - mediana)) / desviacion_estandar
coeficiente_de_curtosis = np.mean(((dataset - media) / desviacion_estandar) ** 4)

# Imprimimos las estadísticas
resultados = {
    "Media": media,
    "Mediana": mediana,
    "Varianza": varianza,
    "Desviación Estándar": desviacion_estandar,
    "Mínimo": minimo,
    "Máximo": maximo,
    "Rango": rango,
    "Percentil 25": percentil_25,
    "Percentil 75": percentil_75,
    "Rango Intercuartílico": rango_intercuartilico,
    "Coeficiente de Variación": coeficiente_de_variacion,
    "Skewness": skewness,
    "Kurtosis": kurtosis,
    "Entropía": entropia,
    "Coeficiente de Asimetría": coeficiente_de_asimetria,
    "Coeficiente de Curtosis": coeficiente_de_curtosis
}

resultados
````

````python
from sklearn import datasets
import pandas as pd
from scipy import stats
import numpy as np

# Cargar el conjunto de datos de Iris de scikit-learn
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                       columns=iris['feature_names'] + ['target'])

# Obtener las variables 'sepal length (cm)' y 'petal length (cm)'
x = iris_df['sepal length (cm)']
y = iris_df['petal length (cm)']

# Realizar la prueba de normalidad Anderson-Darling para ambas variables
anderson_x = stats.anderson(x)
anderson_y = stats.anderson(y)

# Nivel de significancia típico del 5%
significance_level = 5  # Anderson-Darling devuelve niveles de significancia como porcentajes enteros

# Encuentra el índice del valor crítico más cercano al nivel de significancia del 5%
idx_x = np.abs(anderson_x.significance_level - significance_level).argmin()
idx_y = np.abs(anderson_y.significance_level - significance_level).argmin()

critical_value_x = anderson_x.critical_values[idx_x]
critical_value_y = anderson_y.critical_values[idx_y]

# Realizar la prueba de Levene para verificar la homocedasticidad
_, p_value_levene = stats.levene(x, y)

# Realizar la correlación de Pearson para evaluar la relación lineal
pearson_coef, pearson_p = stats.pearsonr(x, y)

# Crear un resumen de los resultados de los tests estadísticos
summary = {
    'Test': ['Anderson-Darling (X)', 'Anderson-Darling (Y)', 'Levene', 'Pearson'],
    'Statistic': [anderson_x.statistic, anderson_y.statistic, None, pearson_coef],
    'P-valor': [None, None, p_value_levene, pearson_p],
    'Valor Crítico': [critical_value_x, critical_value_y, None, None],
    'Significativo': [
        anderson_x.statistic < critical_value_x,
        anderson_y.statistic < critical_value_y,
        p_value_levene > 0.05,
        pearson_p < 0.05
    ],
    'Cumple supuesto': [
        'Sí' if anderson_x.statistic < critical_value_x else 'No',
        'Sí' if anderson_y.statistic < critical_value_y else 'No',
        'Sí' if p_value_levene > 0.05 else 'No',
        'Sí' if pearson_p < 0.05 else 'No'
    ],
    'Comentario': [
        'Normalidad asumida' if anderson_x.statistic < critical_value_x else 'Normalidad no asumida',
        'Normalidad asumida' if anderson_y.statistic < critical_value_y else 'Normalidad no asumida',
        'Homocedasticidad asumida' if p_value_levene > 0.05 else 'Homocedasticidad no asumida',
        'Relación lineal significativa' if pearson_p < 0.05 else 'Relación lineal no significativa'
    ]
}

# Crear DataFrame a partir del resumen para una visualización y análisis más sencillos
summary_df = pd.DataFrame(summary)

# Imprimir el DataFrame resumen con los comentarios
print(summary_df.to_string(index=False))
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression


# Configurar el tema de seaborn para gráficos estilizados
sns.set(style="darkgrid", palette="muted")

# Ajustar un modelo lineal
model = LinearRegression().fit(x[:, np.newaxis], y)
predictions = model.predict(x[:, np.newaxis])
residuals = y - predictions

# Gráfico de residuos para verificar la linealidad y homocedasticidad
plt.figure(figsize=(10, 5))
plt.scatter(x, residuals)  # Changed color
plt.axhline(0, color='darkred', linestyle='--')  # Changed color
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()

# Gráficos Q-Q para 'sepal length (cm)' y 'petal length (cm)'
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

sm.qqplot(x, line='s', ax=ax[0])
ax[0].set_title('Q-Q Plot for Sepal Length (cm)')

sm.qqplot(y, line='s', ax=ax[1])
ax[1].set_title('Q-Q Plot for Petal Length (cm)')

plt.show()

# Crear el gráfico de dispersión con ajuste de regresión
# Ajustamos el color aquí también
joint_plot = sns.jointplot(x=x, y=y, kind='reg', height=8, space=0, joint_kws={'line_kws':{'color':'darkred'}})

# Añadir un título al gráfico usando plt.subplots_adjust y fig.suptitle
joint_plot.fig.subplots_adjust(top=0.9)  # Reduce el espacio superior para hacer espacio para el título
joint_plot.fig.suptitle('Diagrama de Dispersión con Línea de Regresión', fontsize=16)

# Mostrar el gráfico
plt.show()

````
